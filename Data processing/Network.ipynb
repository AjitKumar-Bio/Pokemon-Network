## Template-Matching Color

```python
import networkx as nx
import pickle

# Load the similarity data from file
with open('similarity.txt', 'r', encoding='ANSI') as f:
    similarity_data = [line.strip().split(',') for line in f.readlines()]


# Create a new empty graph
G = nx.Graph()

# Add nodes for each Pokemon
for pokemon in set([d[0] for d in similarity_data]):
    G.add_node(pokemon, Name=pokemon)  # Define the 'Name' attribute for the node

# Add edges between similar Pokemon
for data in similarity_data:
    if float(data[2]) > 0.1:
        pokemon1 = data[0]
        pokemon2 = data[1]
        if not G.has_edge(pokemon1, pokemon2):
            G.add_edge(pokemon1, pokemon2, weight=float(data[2]))

# Save the graph to a file
with open('pokemon_network_color.pickle', 'wb') as f:
    pickle.dump(G, f)

```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Load the graph from file
with open('pokemon_network_color.pickle', 'rb') as f:
    G = pickle.load(f)

# Draw the graph using spring layout
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True)

# Set the edge labels
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

# Show the plot
plt.show()
```
